```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
plt.rcParams['figure.figsize'] = [15, 8]
```

```python
def fun(x, y):
    """ return the function output of f(x,y)= 150sin(ùë•ùë¶/10)‚àí2ùë¶‚àí3ùë•+ùë•ùë¶+ùë•^2+ùë¶^2"""
    return 150*np.sin(x*y/10) -2*y -3*x +x*y + x**2 + y**2
```

Answers Q2:

a) $\nabla (f) = [\frac{\partial f}{\partial x}  \frac{\partial f}{\partial y}]^{T}$

$\frac{\partial f}{\partial x} = 15x\cos(xy/10) -3 + y+2x$

$\frac{\partial f}{\partial y} = 15y\cos(xy/10) -2 + x+2y$

```python
def gradient(x, y):
    """ return the graidnet of the function above."""
    dfdx = 15*x*np.cos(x*y/10) - 3 + y + 2*x
    dfdy = 15*y*np.cos(x*y/10) - 2+ x + 2*y
    return np.array([dfdx, dfdy])
```

b) 

```python
 gradient(-0.6, 7)
```

$\nabla (f(-0.6, 7)) = [-5.41780046,  107.27433873]^{T}$

```python
fun(-0.6, 7)
```

```python
xk = np.array([-0.6, 7])
alpha = 0.01
xk = xk + alpha*gradient(-0.6, 7)
```

```python
print("The new position after one update is = ", xk)
```
